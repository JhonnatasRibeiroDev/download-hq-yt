```python
import os
import requests

# Configura√ß√µes do download
capitulo_inicial_real = 94  # N√∫mero real do cap√≠tulo
capitulo_final_real = 110   # √öltimo cap√≠tulo desejado
codigo_inicial_url = 24948  # C√≥digo do primeiro cap√≠tulo no URL
output_dir = "downloads"    # Pasta onde as imagens ser√£o salvas

# Criar a pasta de sa√≠da se n√£o existir
os.makedirs(output_dir, exist_ok=True)

# C√°lculo do c√≥digo do cap√≠tulo na URL
for capitulo_real in range(capitulo_inicial_real, capitulo_final_real + 1):
    codigo_capitulo_url = codigo_inicial_url + (capitulo_real - capitulo_inicial_real)
    capitulo_dir = os.path.join(output_dir, f"Capitulo_{capitulo_real}")
    os.makedirs(capitulo_dir, exist_ok=True)

    print(f"üìñ Baixando cap√≠tulo {capitulo_real} (C√≥digo: {codigo_capitulo_url})...")

    pagina = 1
    encontrou_pagina = False

    while True:
        pagina_formatada = f"{pagina:04d}"  # Formato 0001, 0002, etc.
        url = f"https://cdn.manhuaplus.cc/ch/{codigo_capitulo_url}/{pagina_formatada}.webp"
        imagem_path = os.path.join(capitulo_dir, f"{pagina_formatada}.webp")

        response = requests.get(url, stream=True)

        if response.status_code == 200:
            encontrou_pagina = True  # Indica que o cap√≠tulo tem pelo menos uma p√°gina
            with open(imagem_path, "wb") as file:
                for chunk in response.iter_content(1024):
                    file.write(chunk)
            print(f"‚úÖ Baixou: {url}")
        else:
            print(f"‚ùå P√°gina {pagina_formatada} n√£o encontrada. Pulando para o pr√≥ximo cap√≠tulo...")
            break  # Sai do loop de p√°ginas

        pagina += 1

    if not encontrou_pagina:
        print(f"üö´ Nenhuma p√°gina encontrada no cap√≠tulo {capitulo_real}. Pulando para o pr√≥ximo...")

print("üéâ Download finalizado!")
```
